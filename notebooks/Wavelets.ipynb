The Haar wavelet can be used to great effect when analysing time series. Using the definition of the Haar, we can constructe the wavelet directl from step functions.

```python
haar = Step(end=0.5,weight=1) - Step(0.5,weight=1)

ninf = Step(0.2)*3

fig, ax = plt.subplots(nrows=3)
x = np.arange(0,1,0.01)
#haar.plot(ax=ax[0])
#ax[0].step(x,haar(x))
prod = haar*ninf
ax[0].step(x,haar(x))

#ninf.plot(ax=ax[1])
ax[1].step(x,ninf(x))
ax[2].step(x,prod(x))
```
