```python
import sys
sys.path.insert(0, r"..//")
import numpy as np
import pandas as pd
from hotstepper.Step import Step
from hotstepper.Steps import Steps
import matplotlib.pyplot as plt
import hotstepper.samples as samples
```

```python
vessel_queue = samples.vessel_queue_sample()
vessel_queue.plot()
```

```python
vessel_queue.ecdf_plot()
```

```python
vessel_queue.summary()
```

```python
#vessel_queue05 = (vessel_queue*Step(end=pd.Timestamp(2020,5,1)))
vc = vessel_queue.clip(ubound=pd.Timestamp(2020,1,10))
vessel_queue05 = vc*0.5
ax = (Steps.apply_reduction_function([vessel_queue05,vc],np.mean)).plot()
vc.plot(ax=ax,color='g')
```

```python
page_views = samples.page_view_sample()
page_views.plot()
page_views_clip = page_views*Step(end=100)
ax = page_views_clip.plot()
Steps.apply_reduction_function([page_views_clip*0.25,page_views],np.min).plot(ax=ax,color='g')
```
